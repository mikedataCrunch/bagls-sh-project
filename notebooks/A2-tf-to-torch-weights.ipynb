## Inspect the `.h5` weights

```python
tf_model = load_model(PATHS.resnet50_weights)
tf_model.summary()
# get weights
weights = tf_model.get_weights()
len(weights)
```

```python
# inspect layers
types = []
names = []
warr_shapes = []
barr_shapes = []
for layer in tf_model.layers:
    types.append(type(layer))
    names.append(layer.get_config()['name'])
    try:
        warr_shapes.append(layer.get_weights()[0])             
        barr_shapes.append(layer.get_weights()[1])
    except:
        warr_shapes.append(layer.get_weights())             
        barr_shapes.append(None)
```

```python
print("Unique layer types: ", set(types))
print("Unique layer names: ", set(names))
```

```python
# param counts for torch
def param_counter_keras(weights):
    counts_per_layer = []
    for warr in weights:
        counts_per_layer.append(np.prod(warr.shape))
    return counts_per_layer
```

```python
keras_param_counts = param_counter_keras(weights)
len(keras_param_counts)
```

## Inspect the `pytorch` model

```python
import torch
```

```python
model = resnet50(weights=None)
```

```python
# param counts for torch
def param_counter_torch(model):
    counts_per_layer = []
    for item in model.parameters():
        counts_per_layer.append(item.numel())
    return counts_per_layer
```

```python
torch_param_counts = param_counter_torch(model)

print("Keras param array length: ", len(keras_param_counts))
print("Torch param array length: ", len(torch_param_counts))
```
