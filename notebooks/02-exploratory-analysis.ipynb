```python
from glob import glob
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.cm as cm

# get random images
def get_sample_paths(src='../test/test', size='all'):
    filepaths = glob(os.path.join(src, '*.png'))
    # exclude segmentation masks
    filepaths = [file for file in filepaths if "_seg" not in file ]
    if size=='all':
        return filepaths
    sample_paths = np.random.choice(filepaths, size=size, replace=False)
    return sample_paths
```

```python
img = mpimg.imread([i for i in get_sample_paths() if "2917" in i][0])
plt.imshow(img)
```

```python
# number of samples to inspect
size = 100

# set up grid
fig, ax = plt.subplots(size // 10, 10, figsize=(20, size//10 * 2.5))
ax = ax.flatten()

for index, path in enumerate(get_sample_paths(size=size)):    
    img = mpimg.imread(path)
    
    fname = path.split('/')[-1]
    dim = img.shape
    
    ax[index].imshow(img, cmap=cm.Greys_r)
    ax[index].set_title(fname + f"\n{dim}", size=8)
    ax[index].axis('off')
plt.show()
```
